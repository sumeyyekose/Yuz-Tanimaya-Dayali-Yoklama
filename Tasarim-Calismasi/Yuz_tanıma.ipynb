```python
import cv2 
import os
import numpy as np
from PIL import Image
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import xlrd, xlwt
from datetime import datetime
from datetime import date



kamera=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')

Id=input('ID Giriniz:')
say=0
while True:

  ret, goruntu= kamera.read()

  gray=cv2.cvtColor(goruntu,cv2.COLOR_BGR2GRAY)
  faces=face_cascade.detectMultiScale(gray,1.3,5)
  for (x,y,w,h) in faces:
    cv2.rectangle(goruntu,(x,y),(x+w,y+h),(255,0,0),2)
    say=say+1
    cv2.imwrite("yuzler/User."+Id +'.'+ str(say) + ".jpg", gray[y:y+h,x:x+w])
    cv2.imshow("sumeyye", goruntu)
  k=cv2.waitKey(100) & 0xFF
  if k==27:
        break
  elif say>=20: 
    break
kamera.release()
cv2.destroyAllWindows()
```

```python
 path = 'yuzler'

tanıma = cv2.face.LBPHFaceRecognizer_create()
face_cascade= cv2.CascadeClassifier("haarcascade_frontalface_default.xml");

def getImagesAndLabels(path): 

    imagePaths=[os.path.join(path,f) for f in os.listdir(path)] 
    #print(imagePaths)

    faceSamples=[]
    Ids=[]

    for imagePath in imagePaths:

        pilImage=Image.open(imagePath).convert('L')
        #print(pilImage)

        goruntu_numpy=np.array(pilImage,'uint8')
        #print(goruntu_numpy)

        
        Id=int(os.path.split(imagePath)[-1].split(".")[1])
        #print(Id)
        faces=face_cascade.detectMultiScale(goruntu_numpy)
        #print(faces)
        for (x,y,w,h) in faces:
            faceSamples.append(goruntu_numpy[y:y+h,x:x+w])
            Ids.append(Id)
    return faceSamples,Ids
#print("eğitim işlemi biraz uzun sürebilir!")
faces,Ids = getImagesAndLabels(path) 
tanıma.train(faces, np.array(Ids))
tanıma.save('egitim/egitim.yml') 
print("\n {0} yüz eğitildi.".format(len(np.unique(Ids))))
```

```python


style_string = "font: bold on; borders: bottom dashed"
style = xlwt.easyxf(style_string)
style1 = xlwt.easyxf(num_format_str='D-MMM-YY')

wb = xlwt.Workbook(encoding="utf-8")
ws = wb.add_sheet('Gelenler', cell_overwrite_ok=True)



df = pd.read_excel('sinif_listesi.xlsx')
ad = df['Ad']
soyad = df['Soyad']
no = df['Id']
sinif = df['Sinif']
uzunluk=len(df)

ws.write(0,4,datetime.now(),style1)
ws.write(0,0,'Id',style=style)
ws.write(0,1,'Ad',style=style)
ws.write(0,2,'Soyad',style=style)
ws.write(0,3,'Sinif',style=style)
for z in range(0,uzunluk):
    ws.write(z+1,0,str(no[z]))
    ws.write(z+1,1,ad[z])
    ws.write(z+1,2,soyad[z])
    ws.write(z+1,3,int(sinif[z]))
    ws.write(z+1,4,"-") 

tanıma = cv2.face.LBPHFaceRecognizer_create() 
tanıma.read('egitim/egitim.yml')
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml" );

cam = cv2.VideoCapture(0) 
font = cv2.FONT_HERSHEY_SIMPLEX 
while True:
   ret, goruntu =cam.read()
   gray=cv2.cvtColor(goruntu,cv2.COLOR_BGR2GRAY)
   faces=faceCascade.detectMultiScale(gray, 1.2,5)
   for(x,y,w,h) in faces:
        cv2.rectangle(goruntu,(x,y),(x+w,y+h),(150,0,0),2)
        
        
        Id, conf = tanıma.predict(gray[y:y+h,x:x+w])

        
        for i in range(0,uzunluk): 
            if Id== no[i]: 
                ws.write(i+1,4,"+",style=style) 

        cv2.putText(goruntu,str(Id),(x+5,y-25), font, 0.5,(0,0,0),1,cv2.LINE_AA)
        cv2.imshow('YUZ TANIMA UYGULAMASI',goruntu);
        

   if(cv2.waitKey(1)==ord('q')):
        wb.save('Yoklama/Gelenler '+str(date.today())+'.xls')
        
        
        
        break;
cam.release()
cv2.destroyAllWindows()
```
